```python
import pandas as pd
import numpy
import sklearn
```

```python
dataset_converted = pd.read_csv("heart.csv")
```

```python
pd.DataFrame(dataset_converted[0:6])
```

```python
X = dataset_converted.loc[:, :'ST_Slope']
pd.DataFrame(X)
```

```python
Y = dataset_converted[["HeartDisease"]]
pd.DataFrame(Y)
```

```python
from sklearn.preprocessing import OneHotEncoder
categorical_columns = dataset_converted[["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]]
```

```python
pd.DataFrame(categorical_columns)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Define column types
numerical_cols = ['Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak']
categorical_cols = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']

# Create preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(drop='first'), categorical_cols)
    ])

# Fit and transform
X_preprocessed = preprocessor.fit_transform(X)
```

```python
pd.DataFrame(X_preprocessed)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, Y, test_size = 0.2, random_state = 42)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.05, 0.1, 0.3],
    'max_depth': [3, 5, 7, 9],
    'min_child_weight': [1, 3, 5],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.3]
}

xgb_model = XGBClassifier(random_state=42, eval_metric='logloss')
```

```python
grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)
```

```python
grid_search.fit(X_train, y_train.values.ravel())
```

```python
#Parameter extraction

print("Best Parameters:", grid_search.best_params_)
print("Best CV Score:", grid_search.best_score_)
```

```python
#predictions
best_xgb_model = grid_search.best_estimator_
y_pred = best_xgb_model.predict(X_test)
```

```python
#Model Eval

print(f"\nTest Accuracy: {accuracy_score(y_test, y_pred):.3f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
#Save Model
import pickle
import joblib
import os

# Get the current directory
current_dir = os.getcwd()
print(f"Saving to: {current_dir}")

# Save XGBoost model
with open('heart_disease_xgb_model.pkl', 'wb') as f:
    pickle.dump(best_xgb_model, f)
print("✓ Model saved as pickle")

# Save XGBoost model as JSON format
best_xgb_model.save_model('heart_disease_xgb_model.json')
print("✓ Model saved as JSON")

# Save the preprocessor
joblib.dump(preprocessor, 'preprocessor.pkl')
print("✓ Preprocessor saved")
```
